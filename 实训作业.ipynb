```python
#（1）导入matplotlib.pyplot模块
from matplotlib import pyplot as plt
#（2）设置中文字体为黑体、中文状态下负号正常显示
plt.rcParams['font.family'] = 'Simhei'
plt.rcParams['axes.unicode_minus'] = False
#（3）根据已知条件设置x轴、y轴数据
x = ['2019年', '2020年', '2021年', '2022年', '2023年']
y1 = [46.4, 46.4, 53.1, 55, 58.6]
y2 = [5.4, 7.2, 9.7, 9.8, 12.6]
y3 = [3.0, 4.3, 6.7, 7.4, 9.4]
#（4）绘制折线图
plt.plot(x, y1, linewidth = 2, color = 'green', linestyle = '--', marker = '*', label = '销售毛利率%')
plt.plot(x, y2, linewidth = 2, color = 'red',  marker = '*', label = '销售净利率%')
plt.plot(x, y3, linewidth = 2, color = 'yellow',  marker = '*', label = '总资产净利率%')
#（5）设置组合折线图标题
plt.title('2017-2021年波司登销售毛利率/销售净利率/总资产净利率')
#（6）显示图例
plt.legend()
#（7）显示数据标签
for a,b in zip(x,y1):
    plt.text(a,b,b,ha='center',va='bottom')
for a,b in zip(x,y2):
    plt.text(a,b,b,ha='center',va='bottom')
for a,b in zip(x,y3):
    plt.text(a,b,b,ha='center',va='bottom')
#（8）显示组合折线图
plt.show()
```
